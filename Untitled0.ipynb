```python
#Показать максимальное median_house_value, где median_income = 3.1250
print (df.loc[(df.median_income == 3.1250), ['median_house_value']].max())
```

```python
#Узнать какая максимальная population в зоне минимального значения median_house_value

df1 = df.loc[df.median_house_value < df.median_house_value.quantile(.25)]
print (df1.population.max())
```

```python
#Задача 40: Работать с файлом california_housing_train.csv, который находится в папке
#sample_data. Определить среднюю стоимость дома, где кол-во людей от 0 до 500 (population)

df1 = df.loc[(df['population']> 0) & (df['population']< 500)][['median_house_value']]
print(df1)
print(df1.median_house_value.mean())
```

```python
#Задача 42: Узнать какая максимальная households в зоне минимального значения population
df1 = df.loc[df.population < df.population.quantile(.20)]
print (df1.households.max())
```
